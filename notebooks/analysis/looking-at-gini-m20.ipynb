```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
data_folder = 'C:/Users/oryan/Documents/galaxy-zoo-desi'
```

```python
df = pd.read_csv(f'{data_folder}/results/gini-m20s.csv', index_col = 0)
```

```python
fig, ax1 = plt.subplots()

df.plot.scatter(
    x = 'm20',
    y = 'gini',
    s = 2.5,
    color = 'black',
    figsize = (5,5),
    title = f'Major Dist. Galaxies (N = {len(df)})',
    ax = ax1
)
ax1.invert_xaxis()
ax1.set_ylabel('Gini (R)')
ax1.set_xlabel('M20 (R)')
plt.show()
```
